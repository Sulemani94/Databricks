**Basic Inteactive SQL Queries**

```
df.createOrReplaceTempView('temp_table')
```

```
%sql
SELECT Time, avg(CI) as AVG_CI
FROM temp_table
GROUP BY Time
ORDER BY Time
```

```
%sql
SELECT Ethnicity, COUNT(Time) AS total
FROM temp_table
GROUP BY Ethnicity
ORDER BY Ethnicity
```

```
%sql
SELECT   Industry, count(Time) as Total
FROM temp_table
GROUP BY Industry
ORDER BY Industry
```
